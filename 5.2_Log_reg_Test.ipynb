```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
df = pd.read_csv("/content/diabetes.csv")
```

```python
df.head()
```

```python
plt.scatter(df.Glucose, df.Outcome, df.BloodPressure, marker = '+', color = 'r')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.view_init(30, 10)
ax.plot3D(df.Glucose, df.Insulin, df.Outcome, 'gray')
```

```python
x = df.drop(columns = 'Outcome', axis=1)
y = df['Outcome']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.23, random_state = 4)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(xtrain, ytrain)
```

```python
xtest_predict = model.predict(xtest)
xtest_predict
```

```python
xtrain_predict = model.predict(xtrain)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest, xtest_predict)
```

```python
accuracy_score(ytrain, xtrain_predict)
```
